```python
from itertools import chain
import random

def play(mazzo: list[tuple[int, str]]):
    random.shuffle(mazzo)
    for i, (v, _) in enumerate(mazzo):
        if v <= 3 and v % 3 == i % 3:
            return False
    return True

results: list[int] = []
for _ in range(1000):
    mazzo = [*chain.from_iterable([[(v, s) for v in range(1, 11)] for s in "CDBS"])]
    N = 10_000
    out = sum([play(mazzo) for _ in range(N)])
    results.append(out)
```

```python
from matplotlib import pyplot as plt
print()
print(avg := sum(results)/len(results), (sum([(x - avg)**2 for x in results])/(len(results)-1))**.5)
plt.hist(results)
```
